# Fórmula general

La fórmula general para resolver ecuaciones cuadráticas ax^2 + bx + c = 0 es:

$$x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$$


```python
import math

def formula_general(a,b,c):
    disc = b*b - 4*a*c
    if disc < 0:
        r1 = complex(-b, math.sqrt(-disc)) / (2*a)
        r2 = complex(-b, -math.sqrt(-disc)) / (2*a)
    else:
        sqrt_d = math.sqrt(disc)
        r1 = (-b + sqrt_d) / (2*a)
        r2 = (-b - sqrt_d) / (2*a)
    return r1, r2

# Ejemplo
print(formula_general(1,-3,2))  # raíces 2 y 1
```
