Introducción a la catálisis
Definición de catálisis
La catálisis es el proceso mediante el cual una sustancia llamada catalizador aumenta la velocidad de una reacción química o permite que ocurra bajo condiciones más suaves (menor temperatura o presión).
Historia de la catálisis
La historia de la catálisis inicia desde lo ancestral hasta lo actual demostrando, así como ha permitido el desarrollo de muchos ámbitos de la sociedad, tanto en la optimización de procesos como en un menor daño ambiental, convirtiéndose en una disciplina científica fundamental. 
Importancia de la catálisis en la industria y procesos biológicos
La catálisis impulsa la sostenibilidad industrial (combustibles limpios, síntesis selectiva, menor consumo de energía) y es esencial para procesos biológicos vitales (enzimas en metabolismo, respiración celular), garantizando eficiencia y ahorro de energía en todos los seres vivos, siendo la base de la vida misma 
En la industria más del 90% de los procesos químicos industriales incorporan catalizadores y cerca del 20% de todos los productos manufacturados se generan mediante procesos catalíticos, como amoniaco, polímeros, combustibles y fármacos. A su vez, se debe tener en cuenta la labor que cumplen los catalizadores automotrices para reducir NOx, CO y HC. 
Diferencia entre catalizadores homogéneos, heterogéneos y biocatalizadores
Se considera catálisis homogénea aquella que comprende sistemas en los que el catalizador y los reactivos se encuentran en una sola fase, normalmente en solución que tiene lugar en una fase única ya sea gas o líquido, más del 90% de los procesos que se realizan en la industria química son catalizados por la combinación de catalizadores homogéneos, los de riesgo medio en estos procesos son su reciclaje y coste con respecto a los procesos heterogéneos (AIMPLAS, 2024).
La catálisis heterogénea tiene lugar cuando el catalizador se encuentra en una fase diferente de la que forman los reactivos y productos, ejerce una influencia decisiva en la cinética de la reacción y en su mecanismo proporciona alto control sobre el lugar y la forma de contacto entre reactivo y catalizador, con el doble objetivo de aumentar la eficiencia del catalizador y disminuir las reacciones no deseadas. En estos procesos, los reactivos entran en contacto con la superficie activa del catalizador, siendo así atravesado por un flujo continuo de reactivos, la influencia de la velocidad de flujo difiere de la catálisis homogénea y la superficie se convierte en un factor significativo: cuanto mayor es el número de sitios activos por unidad de masa dada y mayor la actividad del catalizador (AIMPLAS, 2024).
La catálisis enzimática los catalizadores empleados son enzimas, de naturaleza proteica, que permiten reacciones químicas que, de otro modo, serían extremadamente lentas o no ocurrirían bajo condiciones fisiológicas suaves (temperatura y pH moderados). Las enzimas aceleran las reacciones mediante la disminución de la energía de activación y la estabilización del estado de transición, pudiendo estar inmovilizadas sobre soportes sólidos, como polímeros o matrices inorgánicas, para facilitar su recuperación y reutilización, los procesos que utilizan enzimas inmovilizadas se pueden clasificar como catálisis heterogénea, ya que la enzima inmovilizada constituye una fase distinta a la del medio de reacción, siendo utilizados en síntesis orgánicas de alta selectividad, la catálisis enzimática se destaca por su alta especificidad y eficiencia, generando menos subproductos y residuos, además son fácilmente recuperables y reutilizables (AIMPLAS, 2024).

Propiedades generales de un catalizador
Un catalizador debe cumplir (Fogler, 2016):
Aumentar la velocidad de la reacción.
No consumirse durante la reacción.
No modificar el equilibrio químico (solo acelera la llegada al equilibrio).
Presentar selectividad, permitiendo obtener productos deseados.
Tener estabilidad en las condiciones del proceso.
En catálisis heterogénea, debe tener gran área superficial activa.

Tipos de catálisis
Catálisis ácida: La catálisis ácida es un proceso químico donde un catalizador ácido de tipo Brönsted o Lewis modifica la velocidad de una reacción química, en el cual el catalizador se une a un reactante adecuadamente sustituyendo a otro componente favoreciendo el rompimiento de enlaces y la formación de nuevos productos (Facultad de Ingeniería Química – UNL, 2010).
Catálisis básica: La catálisis básica es un proceso químico utilizado principalmente para aumentar la velocidad de una reacción, en la cual los oxidrilos o bases forman por medio de un sustrato/solvente el producto deseado, destacando algunos procesos industriales como queratolíticos utilizados en la eliminación de durezas en la piel y la producción de biodiésel (Kilpatrick, 1954). 
Catálisis metálica: Los catalizadores metálicos se definen como sustancias transportando moléculas de determinado elemento metálico que alteran la velocidad de reacción de una reacción, siendo un proceso ampliamente utilizado para la producción mediante reactores químicos en la producción petroquímica, donde se puede identificar que la reactividad del catalizador es inherentemente dependiente del metal, el tamaño del cristal y su estado en las condiciones de reacción (Ross, 2019)
Catálisis enzimática: La catálisis enzimática es un proceso catalítico dentro de los sistemas biológicos donde se alteran las velocidades de reacción por la introducción de una enzima, esto mediante sitios específicos donde se adapten y exista afinidad por los reactivos, disminuyendo así la energía de activación de la reacción sin alterar la energía libre del proceso, permitiendo que esta ocurra de forma más eficiente y selectiva (Blanco & Blanco, 2022)

Bases preliminares 
Principios de la termodinámica aplicados a la catálisis
La catálisis no modifica la termodinámica de una reacción:
El cambio en la Energía Libre de Gibbs (ΔG), el cambio en la entalpía (ΔH) y la entropía ΔS permanecen iguales.
El catalizador no cambia el equilibrio químico, solo acelera la velocidad para alcanzarlo más rápido.
La reacción catalizada y la no catalizada tienen la misma energía libre de reacción y el mismo estado de equilibrio.
Principios clave incluyen la Energía Libre de Gibbs, que determina la espontaneidad de la reacción, y el Principio de Sabatier, que establece que un catalizador óptimo tiene una energía de adsorción intermedia con los reactivos: si el enlace es demasiado débil, los reactivos no se adsorben; si es demasiado fuerte, los productos no se desorben fácilmente. Esto da lugar a la clásica relación en forma de "volcán" en catálisis, donde la actividad catalítica es máxima para valores intermedios de energía de adsorción (Masel, 2001). Esta relación conceptual es clave para entender el diseño racional de catalizadores modernos, especialmente en electrocatálisis y procesos de energía limpia, donde la modulación precisa de la energía libre de adsorción en la superficie catalítica define la eficiencia del proceso (Ooka et al., 2021).
El efecto termodinámico clave es que un catalizador proporciona rutas alternativas al estado de transición con menor energía libre de activación, lo que se puede visualizar en diagramas de energía donde la barrera para la reacción catalizada es menor que para la reacción no catalizada, pero ambos comparten los mismos niveles de energía de reactivos y productos. En consecuencia, la catálisis se define más por cambios en la cinética de reacción que en la termodinámica del sistema (Ooka et al., 2021).
Energía de activación y su relación con la velocidad de reacción 
La energía de activación se define como la cantidad mínima de energía necesaria para iniciar una reacción química. Es un parámetro crucial que influye en la velocidad de reacción al controlar la probabilidad de que las moléculas reactivas colisionen con la energía suficiente para experimentar una transformación (Brewster, 2003).
En otras palabras, la energía de activación (Ea) es la energía mínima requerida para que los reactivos formen productos.
El catalizador disminuye la energía de activación, lo que incrementa la constante de velocidad (k) según la ecuación de Arrhenius:
$$k = A e^{-Ea/RT}$$

Una disminución en la energía de activación aumenta exponencialmente la velocidad.

Teoría de los complejos activados 
La Teoría del Complejo Activado (o Teoría del Estado de Transición) explica que para que ocurra una reacción química los reactantes deben chocar y formar una estructura intermedia, antes de formar el producto, de muy alta energía y corta vida llamada complejo activado (o estado de transición). El catalizador facilita la formación de este complejo reduciendo su energía.
Esta teoría proporciona una explicación para la velocidad de una reacción química, es decir, expresa la velocidad de reacción en términos del número de complejos activados (estructuras transicionales en el punto de energía máxima a lo largo de la ruta de reacción) que alcanzan el estado activado para formar los productos.
De acuerdo con esta teoría (Eyring, 1935):
Las moléculas de reactivo forman una especie intermedia llamada complejo activado o estado de transición antes de formar el producto. Este complejo activado es una entidad con alta carga energética y no es aislable.
El complejo activado puede descomponerse para dar los productos o puede desintegrarse para devolver los reactivos.
La diferencia de energía entre los reactivos y el complejo activado se llama energía de activación.
La fracción de moléculas que tienen energía igual o mayor que la energía de activación puede cruzar la barrera energética y formar el producto. Esta fracción de moléculas está dada por la ley de distribución de Boltzmann.
De acuerdo con esta teoría la velocidad depende de la concentración del complejo activado:
$$k = \frac{k_B T}{h} e^{-\Delta G^*/RT}$$

donde:
kB es la constante de Boltzmann
T es la temperatura absoluta
h es la constante de Planck
ΔG* es el cambio de energía de Gibbs del estado de transición
R es la constante universal de los gases
Concepto de equilibrio químico en presencia de catalizadores
Un catalizador es una sustancia que aumenta la velocidad de una reacción química sin consumirse en el proceso al proporcionar una ruta alternativa con menor energía de activación, pero no altera la constante de equilibrio ni la energía de los reactivos o productos.
El catalizador ayuda a que la reacción alcance el equilibrio más rápidamente acelerando tanto la reacción directa como la inversa sin afectar la composición del equilibrio o los estados energéticos inicial y final. Como se indica en la literatura clásica de equilibrio químico, el papel del catalizador es puramente cinético, ya que no altera el valor de la constante de equilibrio K, la cual depende únicamente de la temperatura y de las energías libres de los estados inicial y final (Denbigh, 1981).
Principio de Le Chatelier aplicado a reacciones catalizadas 
El Principio de Le Chatelier describe cómo un sistema en equilibrio responde ante cambios externos (como temperatura, presión o concentración) para minimizar dichos cambios y restablecer el equilibrio mediante el ajuste de la posición de equilibrio, mientras que la introducción de un catalizador en una reacción en equilibrio no modifica la posición de equilibrio ni el valor de K, sino que acelera tanto la reacción directa como la inversa facilitando el ajuste dinámico ante estos cambios externos, sin modificar las cantidades finales de reactivos y productos.
Ejemplo: Si una reacción exotérmica se cataliza, el catalizador no evita que el aumento de temperatura favorezca la reacción inversa, según Le Chatelier, pero sí hace que el nuevo equilibrio se alcance más rápido. Esto explica cómo, en sistemas complejos (como redes bioquímicas o procesos sensoriales), la combinación de catálisis y principios de equilibrio permite respuestas rápidas pero reguladas a perturbaciones externas (Norwich, 2010).
Factores que afectan la actividad catalítica (temperatura, presión, concentración, etc.)
Los factores clave que afectan la actividad catalítica de enzimas y otros catalizadores incluyen la temperatura, el pH, la concentración de reactivos y la presencia de inhibidores o activadores. Estos parámetros determinan tanto la afinidad del catalizador por el sustrato como la velocidad de formación del estado de transición, siendo críticos para mantener una catálisis eficiente, con puntos óptimos específicos para cada enzima. 
Factores principales
Temperatura:
Bajas temperaturas: Reducen la energía cinética, disminuyendo colisiones enzima-sustrato y velocidad de reacción.
Temperatura Óptima: Máxima actividad.
Temperaturas Altas:  la enzima (cambian su forma), perdiendo función catalítica.
pH:
Cada enzima tiene un pH óptimo, donde la estructura del sitio activo está en condiciones ideales para la unión del sustrato.
pH extremos pueden protonar o desprotonar grupos funcionales críticos, alterando la forma y carga del sitio activo.
Concentración de sustrato:
Aumentar el sustrato incrementa la velocidad inicialmente.
Se alcanza un punto de saturación donde la velocidad se estabiliza, porque todas las enzimas están ocupadas.
Inhibidores y Activadores:
Inhibidores: Sustancias que disminuyen o detienen la actividad enzimática (ej., venenos, algunos fármacos).
Activadores (Cofactores/Coenzimas): Iones metálicos o moléculas orgánicas que se requieren para que la enzima sea activa y funcione correctamente. 
Los catalizadores diferentes a las enzimas también se ven afectados por la presencia de impurezas o venenos, como azufre o compuestos que generan coque, especialmente en catalizadores metálicos o soportados.
Además, las velocidades de reacción comúnmente dependen de la concentración de reactivo, la temperatura y la superficie de catalizador líquido o sólido, y en aplicaciones específicas, la velocidad depende de factores adicionales, como la topografía superficial como el equilibrio de adsorción–desorción.


# 3. MECANISMOS DE ACCIÓN CATALÍTICA



## 3.1. Mecanismos generales de acción catalítica

La catálisis consiste en acelerar una reacción química mediante la disminución de la energía de activación necesaria para que los reactivos se transformen en productos. Desde un punto de vista mecanístico, el catalizador ofrece una ruta alternativa a la reacción no catalizada, en la cual el estado de transición tiene una energía más baja.

En términos mecanísticos, el catalizador suele unirse de manera transitoria a uno o más reactivos, formando especies intermedias. Estas especies pueden ser complejos adsorbidos en la superficie (en el caso de catálisis heterogénea) o complejos de coordinación en solución (en catálisis homogénea). Durante el ciclo catalítico, el catalizador participa en varios pasos elementales, como adsorción, reacción superficial, desorción o transferencia de electrones, pero al final del proceso recupera su estructura inicial.

Es importante resaltar que la acción catalítica solo afecta a la cinética de la reacción y no a su termodinámica: no modifica la energía libre de Gibbs de la reacción global ni la posición del equilibrio, de modo que tanto la reacción directa como la inversa se ven aceleradas en la misma proporción.



## 3.2. Teoría del complejo de transición y su relación con los mecanismos de reacción

La Teoría del Complejo de Transición, también conocida como Teoría del Estado de Transición, proporciona un marco conceptual para entender cómo se forman y rompen los enlaces químicos durante una reacción. De acuerdo con esta teoría, los reactivos deben pasar por un estado de alta energía, el complejo activado, antes de transformarse en productos.

En catálisis, el catalizador estabiliza este complejo de transición al interactuar con los reactivos, reduciendo la barrera energética que separa reactivos y productos. Esto puede lograrse mediante:

- Formación de enlaces débiles (por ejemplo, interacciones de Van der Waals o enlaces de hidrógeno) que facilitan la reorganización electrónica.
- Coordinación de los reactivos a centros metálicos en catalizadores de transición, lo que permite una redistribución controlada de densidad electrónica.
- Provisión de sitios activos con geometría y polaridad adecuadas para estabilizar el estado de transición.

El análisis detallado del perfil de energía potencial y de los estados de transición involucrados permite proponer mecanismos de reacción, identificar pasos determinantes de la velocidad y diseñar catalizadores más eficientes.



## 3.3. Mecanismos de catálisis en sistemas homogéneos y heterogéneos

En **catálisis homogénea**, el catalizador y los reactivos se encuentran en la misma fase (habitualmente en solución). Los mecanismos típicos incluyen:

- **Mecanismo de asociación**: el catalizador forma primero un complejo con el reactivo, que posteriormente se transforma en producto.
- **Mecanismo de disociación**: el catalizador genera un intermediario reactivo que luego reacciona con el sustrato libre.
- **Ciclos catalíticos organometálicos**: con etapas de adición oxidativa, inserción y eliminación reductora.

En **catálisis heterogénea**, el catalizador está en una fase distinta (generalmente sólido) y los reactivos en fase gas o líquido. Los mecanismos generales comprenden:

1. **Difusión externa**: transporte del reactivo desde el fluido a la superficie del catalizador.
2. **Adsorción**: fijación del reactivo en la superficie activa (física o química).
3. **Reacción superficial**: transformación química sobre los sitios activos.
4. **Desorción**: liberación de los productos desde la superficie al fluido.
5. **Difusión interna** (si el catalizador es poroso): transporte dentro de los poros.

El paso determinante de la velocidad puede situarse en cualquiera de estas etapas, y su identificación es clave para el diseño y optimización de reactores catalíticos.



## 3.4. Catálisis por contacto o por adsorción

La catálisis por contacto es un caso particular de catálisis heterogénea en la que la reacción ocurre casi exclusivamente en la superficie del catalizador. La adsorción juega un papel central y puede ser de dos tipos:

- **Fisisorción**: interacción débil (fuerzas de Van der Waals), reversible y con baja energía de adsorción.
- **Quimisorción**: formación de enlaces químicos entre el reactivo y la superficie, con energías de adsorción mayores.

En la catálisis por adsorción, el mecanismo básico implica:

1. Adsorción de los reactivos en sitios activos.
2. Activación y ruptura/formación de enlaces sobre la superficie.
3. Desorción de los productos.

La naturaleza química y estructural de la superficie (composición, facetas cristalinas, defectos, etc.) controla la selectividad y actividad del catalizador.



## 3.5. Ciclo de Langmuir–Hinshelwood

El mecanismo de **Langmuir–Hinshelwood** describe reacciones catalíticas superficiales en las que **ambos reactivos** se adsorben en la superficie del catalizador antes de reaccionar entre sí. Los pasos típicos son:

1. Adsorción de A en la superficie.
2. Adsorción de B en la superficie.
3. Reacción entre A* y B* (especies adsorbidas) para formar un producto P*.
4. Desorción de P hacia la fase fluida.

Este modelo asume una monocapa de adsorción y sitios activos equivalentes, y conduce a ecuaciones de velocidad que dependen de las coberturas de superficie y de las constantes de adsorción y reacción. Es muy utilizado para describir reacciones en catálisis heterogénea, como oxidaciones parciales, hidrogenaciones y reacciones sobre metales de transición.



## 3.6. Catálisis enzimática: mecanismos específicos

En la **catálisis enzimática**, el mecanismo general puede representarse mediante el modelo de Michaelis–Menten:

E + S ⇌ ES → E + P

donde E es la enzima, S el sustrato, ES el complejo enzima–sustrato y P el producto.

Los mecanismos específicos incluyen:

- **Catálisis ácido–base**: transferencia de protones desde o hacia grupos funcionales del sustrato.
- **Catálisis covalente**: formación de enlaces transitorios entre la enzima y el sustrato.
- **Catálisis por iones metálicos**: utilización de cofactores metálicos para estabilizar estados de transición o facilitar transferencias electrónicas.

La complementariedad estructural entre el sitio activo y el sustrato, junto con la dinámica conformacional de la proteína, es fundamental para la alta especificidad y eficiencia observadas en las enzimas.



## 3.7. Modelos de sitios activos y su interacción con los reactivos

Los sitios activos en un catalizador representan las regiones donde la adsorción y transformación química son más favorables. Estos sitios pueden ser:

- Átomos de superficie coordinativamente insaturados.
- Defectos cristalinos (escalones, bordes, esquinas).
- Centros metálicos en estructuras organometálicas.
- Grupos funcionales específicos en catalizadores ácidos o básicos.

La interacción con los reactivos depende de la naturaleza electrónica y geométrica del sitio activo. La teoría de enlaces de quimisorción, el modelo d-band y los cálculos de teoría del funcional de la densidad (DFT) permiten racionalizar estas interacciones y predecir tendencias en actividad y selectividad.



## 3.8. Key–Lock

El modelo de **llave–cerradura (Key–Lock)** propuesto inicialmente por Emil Fischer describe la especificidad enzimática suponiendo que el sitio activo de la enzima tiene una forma complementaria al sustrato, de manera análoga a una llave que encaja exactamente en una cerradura.

Aunque este modelo explica de manera sencilla la especificidad, no considera los cambios conformacionales de la enzima ni la flexibilidad del sitio activo, por lo que ha sido ampliado por modelos más modernos como el de ajuste inducido.



## 3.9. Modelo de ajuste inducido

El modelo de **ajuste inducido** propone que el sitio activo de la enzima no es completamente complementario al sustrato en su estado libre, sino que sufre cambios conformacionales al unirse a este. La interacción inicial induce una reorganización estructural que optimiza el acoplamiento enzima–sustrato y estabiliza el estado de transición.

Este modelo explica mejor la alta especificidad y afinidad observadas en muchas enzimas, así como fenómenos de regulación alostérica y cooperatividad. Además, resalta la importancia de la dinámica y flexibilidad de las biomoléculas en los procesos catalíticos.

Mecanismos de degradación catalítica

Saturación
La saturación ocurre cuando todos los sitios activos del catalizador se encuentran ocupados por moléculas de reactivo, de modo que un aumento ulterior en la concentración de reactivo no incrementa la velocidad (Cornish-Bowden, 2012).
En catálisis enzimática se relaciona con la constante Km (modelo de Michaelis-Menten), donde la velocidad llega a un valor Vmáx.
Envenenamiento 
El envenenamiento es la pérdida de actividad catalítica debida a la adsorción fuerte de especies que bloquean los sitios activos (Bartholomew, 2001).
Tipos de venenos comunes
Azufre (S): envenena catalizadores metálicos como Ni o Pt.
Fósforo (P): afecta catalizadores de cracking.
Compuestos orgánicos pesados, coque o halógenos.
El veneno reduce la actividad porque los reactivos no pueden adsorberse.
Bloqueo 
El bloqueo ocurre cuando productos, impurezas o depósitos tapan físicamente los poros o la superficie, reduciendo el acceso a los sitios activos. Por ejemplo, en catalizadores de cracking, el coque acumulado en los poros bloquea el acceso de los hidrocarburos; o en catalizadores de hidrogenación, la presencia de azufre sobre un catalizador básico puede bloquear sitios activos.
Definición y tipos de degradación catalítica 
La degradación catalítica es la pérdida parcial o total de actividad y/o selectividad de un catalizador a lo largo del tiempo, debido a modificaciones físicas o químicas de su estructura y composición, generadas por las condiciones de operación, la presencia de contaminantes o la propia reacción (Bartholomew & Farrauto, 2006).
Principales tipos de degradación catalítica
Envenenamiento: Ocurre cuando especies químicas externas (impurezas, venenos específicos) se adsorben fuertemente sobre los sitios activos del catalizador, bloqueándolos de manera irreversible o semirreversible.
Fouling o bloqueo (ensuciamiento): Consiste en la deposición de sustancias (como coque, polímeros u óxidos) sobre la superficie o dentro de los poros del catalizador, reduciendo el área superficial disponible y obstruyendo el contacto con los reactivos.
Sinterización de catalizadores: El sinterizado de catalizadores es el proceso donde las partículas o cristales catalíticos se agrandan, disminuyendo su área específica y su cantidad de sitios activos, ya sea por fusión parcial o reestructuración superficial a altas temperaturas, provocando una pérdida significativa de actividad y cambios en la selectividad de la reacción (Bartholomew & Farrauto, 2006).
La sinterización de catalizadores, especialmente en sistemas metálicos soportados (como Pt/γ-Al2O3 o Ni/SiO2), se ve fuertemente influida por parámetros como la temperatura, la presión parcial de los reactivos y la presencia de compuestos reductores u oxidantes. A lo largo del proceso, las partículas finas tienden a coalescer y formar agregados mayores, disminuyendo el número de sitios activos expuestos. Este fenómeno puede ser también acelerado por fluctuaciones térmicas y atmósferas reactivas, como mezclas ricas en vapor de agua o hidrógeno, que modifican la movilidad superficial de las especies metálicas (Bartholomew, 2001).
Desde el punto de vista termodinámico, el sinterizado está impulsado por la minimización de la energía superficial total del sistema, lo que favorece la formación de partículas más grandes con menor área específica por unidad de volumen. Esto tiene consecuencias directas en la actividad catalítica, especialmente en reacciones donde la velocidad global depende fuertemente del número de sitios activos expuestos, como en procesos de hidrogenación, deshidrogenación y reformado (Bartholomew & Farrauto, 2006).
Para reducir el impacto de la sinterización, se han desarrollado estrategias basadas en la estabilización estructural del catalizador. Estas incluyen el uso de soportes de alta estabilidad térmica (por ejemplo, óxidos mixtos como CeO2–ZrO2), la incorporación de promotores que anclan las especies metálicas a la superficie, y el diseño de estructuras tipo núcleo–cáscara que protegen las nanopartículas activas frente a la coalescencia. Adicionalmente, la operación a temperaturas moderadas y el control fino de la atmósfera de reacción (por ejemplo, evitando condiciones fuertemente reductoras a alta temperatura) son esenciales para preservar la dispersión metálica a lo largo del tiempo (Somorjai & Li, 2010).
Desactivación por contaminación 
La desactivación por contaminación en catálisis se refiere a la pérdida progresiva de actividad o selectividad debido a la incorporación de especies extrañas al sistema, generalmente provenientes de impurezas en los reactivos, el medio de reacción o incluso del propio equipo de proceso, estas especies pueden ser compuestos metálicos, sales, partículas sólidas, azufre o fósforo, los cuales actúan obstruyendo sitios activos, modificando la estructura del soporte o generando fases inactivas, reduciendo la eficiencia del catalizador (Bartholomew & Farrauto, 2006).
Desde una perspectiva superficial, la contaminación puede provocar cambios notables en la morfología y composición química de la superficie del catalizador. Por ejemplo, en catalizadores metálicos soportados, la deposición de contaminantes puede alterar la distribución electrónica de las nanopartículas, modificar la energía de adsorción de los reactivos y cambiar la acidez o basicidad del soporte; en el caso de catalizadores heterogéneos, la presencia de partículas extrañas puede obstruir poros, disminuir el área superficial efectiva y modificar la humectabilidad del sólido, afectando la transferencia de masa y la difusión de reactivos y productos a través de la estructura porosa (Somorjai & Li, 2010).
La contaminación también se asocia con la formación de fases secundarias inactivas o poco activas, por ejemplo, la reacción entre el catalizador y ciertos contaminantes puede dar lugar a la formación de sulfuros, fosfatos u óxidos estables, que no participan en la reacción catalítica pero ocupan posiciones clave en la superficie delimitando la disponibilidad del material activo, estos fenómenos son particularmente críticos en procesos de larga duración, como la refinación de hidrocarburos, la hidrodesulfuración o la catálisis ambiental, donde pequeñas cantidades de impurezas pueden, con el tiempo, degradar sustancialmente el rendimiento del catalizador (Bartholomew, 2001).
A nivel macroscópico, la desactivación por contaminación se traduce en una disminución progresiva de la conversión, selectividad o rendimiento del proceso, obligando a operar a condiciones más severas (mayor temperatura o presión) para compensar la pérdida de actividad, esto no solo aumenta los costos operativos, sino que también puede promover mecanismos adicionales de degradación, como sinterización o coqueo acelerado.
En la práctica, la prevención de la desactivación por contaminación requiere el uso de materias primas de alta pureza, sistemas de pretratamiento (filtración, desulfuración, desmetalización), y el diseño de catalizadores más robustos, capaces de tolerar niveles moderados de impurezas sin perder su funcionalidad. Además, la monitorización continua de la composición de las corrientes de alimentación y la implementación de protocolos de regeneración periódica son estrategias clave para prolongar la vida útil del catalizador en sistemas industriales complejos.
Degradación en sistemas heterogéneos: análisis de la estructura superficial)
En sistemas heterogéneos, la degradación de catalizadores está estrechamente ligada a cambios en la estructura superficial, tanto a nivel morfológico como electrónico, por ejemplo, en catalizadores metálicos soportados, la pérdida de área específica activa puede estar asociada a la formación de agregados de mayor tamaño (sinterización), colapso parcial de la estructura del soporte o recubrimiento por depósitos de coque o contaminantes.
Desde una perspectiva estructural, la reorganización superficial puede afectar la distribución y naturaleza de los sitios activos (escalones, bordes, defectos cristalinos), lo cual tiene un impacto directo en la selectividad de la reacción. En muchos casos, las reacciones deseadas se ven favorecidas en sitios específicos, mientras que los sitios generados durante la degradación pueden promover reacciones secundarias no deseadas o menos selectivas.
El análisis de la estructura superficial se realiza mediante técnicas avanzadas como microscopia electrónica de transmisión (TEM), microscopia de fuerza atómica (AFM), difracción de rayos X (XRD), espectroscopías de fotoelectrones (XPS) y espectroscopías de adsorción-desorción de gases (BET), entre otras, estas herramientas permiten correlacionar cambios en la actividad catalítica con modificaciones estructurales a nivel nano y microscópico, proporcionando información crítica para el diseño de catalizadores más resistentes a la degradación.
Consecuencias generales de la degradación superficial
Disminución progresiva de la actividad catalítica.
Cambios en la selectividad, debido a la alteración del tipo y distribución de sitios activos.
Aumento de resistencias difusionales internas.
Reducción de la vida útil del catalizador.
Necesidad de regeneración o reemplazo del material.
En conjunto, estos procesos hacen que la degradación superficial sea un factor clave en la evaluación de la estabilidad y durabilidad de los catalizadores heterogéneos, siendo esencial su estudio para el diseño de materiales con larga vida útil y alta resistencia a condiciones de operación críticas de desactivación en la catálisis heterogénea.
Reacciones secundarias y su impacto en la vida útil del catalizador 
Las reacciones secundarias son procesos no deseados que ocurren simultáneamente con la reacción principal, generando productos alternativos que pueden disminuir el rendimiento, alterar la selectividad y promover la degradación del catalizador aumentando el hábito de que especies de reacción se transformen en subproductos que pueden tener mayor afinidad por la superficie catalítica o por los poros del soporte, lo cual facilita fenómenos de ensuciamiento y bloqueo, un ejemplo típico es la formación de coque en procesos de reformado o craqueo catalítico, donde las reacciones secundarias de polimerización o condensación aromática conducen a especies altamente insaturadas y poliaromáticas que se depositan sobre el catalizador (Moulijn et al., 2001).
Mecanismos de regeneración de catalizadores 
La regeneración de catalizadores es el conjunto de procedimientos mediante los cuales se busca restaurar, total o parcialmente, la actividad, selectividad y estabilidad de un catalizador desactivado, esto se logra mediante la eliminación o conversión de depósitos (como coque, azufre o compuestos metálicos), la reversión de cambios estructurales o la reactivación de sitios activos bloqueados, y la regeneración puede ser in situ, cuando se realiza directamente en el reactor, o ex situ, cuando el catalizador se retira del sistema para ser tratado en unidades especiales, teniendo un papel crucial en procesos industriales continuos, donde reemplazar el catalizador con frecuencia sería económica y técnicamente inviable (Hatta et al., 2023).
Las estrategias de regeneración dependen del tipo de desactivación:
Para **ensuciamiento por coque**, se emplea a menudo una oxidación controlada mediante aire o mezclas de oxígeno y vapor, eliminando el carbono depositado sin destruir la estructura del soporte.
Para **envenenamiento** por azufre o halógenos, puede utilizarse hidrogenación reductora o tratamientos químicos específicos que reactiven los sitios metálicos.
En el caso de **sinterización**, la regeneración es más compleja, ya que implica cambios estructurales cuasi irreversibles, aunque en algunos sistemas es posible redispersar las especies metálicas mediante tratamientos de oxidación–reducción y reimpregnación.
Regeneración en procesos industriales:
En la industria de refino, por ejemplo, los catalizadores de cracking catalítico fluido (FCC) se regeneran continuamente en un regenerator, donde el coque depositado se quema con aire controlado, recuperando la actividad catalítica antes de recircular al reactor. De manera similar, en procesos de hidrotratamiento, los catalizadores se regeneran periódicamente mediante etapas de oxidación e hidrogenación.
Estrategias para prevenir o minimizar la degradación catalítica
La prevención de la degradación catalítica es fundamental para maximizar la vida útil y la eficiencia de los catalizadores en aplicaciones industriales, especialmente en procesos de larga duración, en términos generales, la estrategia más efectiva consiste en una combinación de diseño racional del catalizador, control estricto de las condiciones de operación y manejo adecuado de materias primas y corrientes de proceso, la prevención se aborda desde etapas tempranas de diseño del proceso, considerando no solo la actividad inicial, sino la estabilidad a largo plazo en condiciones reales de operación (Bartholomew & Farrauto, 2006).
En primer lugar, el control adecuado de temperatura y presión es crucial para reducir fenómenos de sinterización, volatilización de componentes activos o colapso estructural del soporte, operar dentro de ventanas térmicas seguras y evitar transitorios bruscos contribuye a mantener la integridad del material catalítico. Asimismo, la selección de atmósferas de reacción adecuadas (por ejemplo, relaciones óptimas de gases oxidantes/reductores) puede minimizar procesos de oxidación o reducción excesivas que conduzcan a la formación de fases inactivas.
Otra estrategia fundamental consiste en minimizar la presencia de impurezas en las corrientes de alimentación, ya que compuestos como azufre, fósforo, metales pesados o cloruros pueden actuar como venenos específicos, para ello, se emplean unidades de pretratamiento (desulfuración, desalado, filtración) y se monitoriza la calidad de las materias primas y corrientes de proceso. Adicionalmente, el uso de promotores y modificadores estructurales en el diseño del catalizador puede aumentar su tolerancia a contaminantes, reforzar la estabilidad de los sitios activos y reducir la tendencia a la formación de coque o depósitos inactivos.
En cuanto al diseño del catalizador, la incorporación de propiedades estructurales robustas y soportes de alta estabilidad térmica (como óxidos mixtos o materiales mesoporosos ordenados) contribuye a mitigar fenómenos de sinterización y colapso de poros, además, la optimización de la distribución de poros y tamaño de partículas permite un equilibrio entre accesibilidad de los reactivos y resistencia a la obstrucción, las formulaciones con fases activas ancladas o encapsuladas también pueden limitar la migración y coalescencia de especies metálicas, reduciendo la degradación superficial.
Finalmente, la implementación periódica de ciclos de regeneración suaves y controlados, así como la monitorización en línea de variables clave (conversión, selectividad, caída de presión, composición de efluentes), permite detectar tempranamente los signos de degradación, ajustando condiciones operativas o aplicando tratamientos regenerativos antes de que la pérdida de actividad sea crítica, en conjunto, estas estrategias permiten prolongar la vida útil del catalizador y mantener su rendimiento óptimo durante el mayor tiempo posible.
Aspectos avanzados y nuevas tendencias en catálisis
Catálisis en la economía circular y sostenibilidad 
La catálisis desempeña un papel fundamental en la economía circular al facilitar procesos químicos más eficientes, selectivos y de menor impacto ambiental, permitiendo el aprovechamiento óptimo de recursos y la reducción de residuos, mediante el uso de catalizadores avanzados es posible transformar desechos o subproductos industriales en materias primas de alto valor añadido, tales como combustibles sintéticos, compuestos químicos finos o materiales funcionales, además, la catálisis posibilita la utilización de fuentes renovables de carbono (como biomasa o CO2 capturado) como sustratos para la síntesis de nuevas moléculas, contribuyendo a la descarbonización de la industria química (Parlamento Europeo, 2023).
Asimismo, la catálisis permite llevar a cabo procesos bajo condiciones más suaves de temperatura y presión, lo que disminuye significativamente el consumo energético y las emisiones asociadas, en el contexto de la transición energética, la catálisis es clave en tecnologías como la producción de hidrógeno verde por electrólisis, la reducción electroquímica de CO2 y la síntesis de combustibles sostenibles a partir de residuos, estas aplicaciones no solo mejoran la eficiencia y selectividad de las transformaciones químicas, sino que también favorecen la integración de energías renovables en la industria.
Avances en catálisis de metales en forma de nanopartículas 
Las nanopartículas metálicas han emergido como una de las plataformas más versátiles en catálisis heterogénea y homogénea debido a su alta relación superficie/volumen y a su capacidad para exhibir propiedades electrónicas y geométricas ajustables, la disminución del tamaño de partícula hasta la escala nanométrica incrementa notablemente el número de sitios activos disponibles y puede modificar la estructura de bandas electrónicas, generando efectos de confinamiento cuántico que influyen en la adsorción y activación de moléculas reactantes (Del Rocío Redón de la Fuente, 2016).
Dentro de los avances recientes, se destacan los catalizadores basados en nanopartículas bimetálicas o multimetálicas, donde la combinación controlada de diferentes metales permite ajustar finamente la actividad y selectividad mediante efectos sinérgicos, por ejemplo, la incorporación de un segundo metal puede mejorar la resistencia a la sinterización, modular la energía de adsorción de especies clave (como H, CO o CO2) o introducir nuevos sitios activos específicos para etapas críticas de una ruta de reacción, además, el uso de soportes avanzados, como materiales mesoporosos ordenados, óxidos mixtos o estructuras tipo MOF, contribuye a estabilizar las nanopartículas y mejorar la dispersión, prolongando la vida útil del catalizador (Romero et al., 2024).
Uso de catalizadores reciclables y regenerables
El desarrollo de catalizadores reciclables y regenerables constituye una de las tendencias más importantes en catálisis sostenible, ya que permite reducir el consumo de materiales críticos (como metales nobles) y minimizar la generación de residuos sólidos, entre las estrategias destacadas se encuentran los catalizadores soportados en materiales magnéticos, que pueden recuperarse fácilmente mediante campos magnéticos, y los catalizadores anclados en fases poliméricas o híbridas, que combinan estabilidad química con facilidad de manejo y separación (Guibelondo, 2025).
Una de las áreas con mayor crecimiento es el desarrollo de sistemas catalíticos que mantengan su actividad tras múltiples ciclos de reacción y regeneración, sin perder actividad. Esto incluye (Guibelondo, 2025):
catalizadores heterogéneos robustos, diseñados para múltiples ciclos de operación,
catalizadores magnéticos, que pueden separarse fácilmente mediante campos magnéticos,
polímeros catalíticos y organocatalizadores recuperables por extracción,
sistemas híbridos orgánico–inorgánicos con estabilidad mejorada.
Además, la regeneración avanzada combina técnicas térmicas, químicas y electroquímicas para restaurar la actividad catalítica sin necesidad de sustituir completamente el material, generando un impacto positivo tanto en la eficiencia económica como en la huella ambiental, favoreciendo procesos industriales más sostenibles.

### 7.1 Catálisis computacional y modelado molecular

La catálisis computacional emplea métodos de química cuántica, teoría del funcional de la densidad (DFT) y simulaciones de dinámica molecular para estudiar la estructura electrónica, los estados de transición y la superficie de energía potencial de sistemas catalíticos complejos. Gracias a estos métodos, se pueden explorar mecanismos de reacción, barreras de activación y sitios activos preferenciales con una resolución que resulta difícil de alcanzar exclusivamente mediante técnicas experimentales.

La simulación por computadora permite predecir la estabilidad de intermedios de reacción, la energía de adsorción de reactivos y productos, y la influencia de modificaciones estructurales del catalizador (como dopaje con otros metales, cambio de soporte o variación de la morfología). De este modo, el modelado molecular se convierte en una herramienta clave para el diseño racional de catalizadores, ya que posibilita evaluar múltiples candidatos de forma virtual antes de su síntesis experimental.

Gracias al modelado molecular es posible predecir la reactividad y selectividad de nuevos materiales, optimizar la composición de catalizadores bimetálicos o multimetálicos, y estudiar entornos reactivos complejos (como interfaces sólido–líquido o electrocatálisis) sin necesidad de sintetizar los catalizadores experimentalmente.



### 7.2 Evaluación catalítica de metales preciosos frente a metales base

La evaluación comparativa entre metales preciosos (como Pt, Pd, Rh, Au) y metales base (como Ni, Co, Cu, Fe) constituye un área central en el desarrollo de catalizadores más económicos y sostenibles. Los metales preciosos suelen exhibir una alta actividad y selectividad en numerosas reacciones (hidrogenación, oxidación selectiva, reformado), además de poseer una notable estabilidad frente a la sinterización y la desactivación por envenenamiento. Sin embargo, su elevado costo y la limitada disponibilidad impulsan la búsqueda de alternativas basadas en metales más abundantes.

Los metales base, por su parte, ofrecen una alternativa más asequible, pero pueden presentar desafíos en cuanto a estabilidad térmica, resistencia a la corrosión o susceptibilidad al envenenamiento por azufre, fósforo u otras impurezas. Por ello, una línea de investigación destacada consiste en modificar estos metales mediante aleaciones, dopajes, soportes avanzados o tratamientos superficiales que mejoren su actividad y durabilidad, acercándolos al desempeño de los metales nobles.

En la práctica, la evaluación catalítica se basa en comparar parámetros como conversión, selectividad, rendimiento, actividad específica (por unidad de superficie o número de sitios activos) y estabilidad a largo plazo. De esta manera, se pueden identificar formulaciones de metales base altamente optimizadas que, en ciertas aplicaciones, compiten directamente con catalizadores basados en metales preciosos.



### 7.3 Catálisis en condiciones extremas

La catálisis en condiciones extremas abarca procesos que operan a temperaturas muy elevadas, presiones extremas, medios altamente corrosivos o en presencia de radiación intensa. Estas condiciones son frecuentes en sectores como la producción de combustibles avanzados, la conversión de recursos no convencionales (como gas de esquisto o arenas bituminosas), y la valorización de residuos.

En estos entornos, los catalizadores convencionales pueden sufrir degradación acelerada, sinterización, lixiviación de especies activas o colapso estructural del soporte. Por ello, se diseñan materiales específicamente adaptados, como óxidos refractarios, carburos, nitruros, fases tipo perovskita o materiales cerámicos avanzados, capaces de mantener su integridad bajo tales exigencias.

Además, el estudio de la catálisis en condiciones extremas se ha extendido a campos como la astroquímica y la geociencia, donde se investigan procesos catalíticos que ocurren en el interior de planetas o en superficies minerales expuestas a ambientes extremos. Estos estudios no solo tienen interés fundamental, sino que también inspiran nuevas soluciones tecnológicas para la industria.



### 7.4 Desarrollo de nuevos materiales catalíticos (nanotecnología)

La nanotecnología ha abierto un amplio abanico de posibilidades en el diseño de **nuevos materiales catalíticos**, ya que permite controlar de manera precisa el tamaño, la forma, la composición y la estructura de los catalizadores a escala nanométrica. Las nanopartículas metálicas, los nanocompuestos y los materiales híbridos orgánico–inorgánicos pueden presentar propiedades catalíticas significativamente distintas a sus análogos macroscópicos.

Entre las estrategias más destacadas se encuentran:

- **Nanopartículas soportadas** sobre óxidos, carbonos porosos o materiales poliméricos.
- **Estructuras núcleo–cáscara**, donde un núcleo activo se recubre con una capa protectora o moduladora.
- **Materiales bidimensionales (2D)** como grafeno, disulfuros de metales de transición y otros análogos, utilizados como soportes conductores y altamente dispersantes.
- **MOFs (Metal–Organic Frameworks)** y COFs, que ofrecen arquitecturas porosas altamente ordenadas y ajustables.

El desarrollo de estos materiales se apoya fuertemente en técnicas avanzadas de caracterización (microscopías electrónicas de alta resolución, espectroscopías in situ y operando) y en herramientas de simulación computacional, que permiten correlacionar estructura, propiedades electrónicas y desempeño catalítico.
